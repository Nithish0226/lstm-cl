```python
from lstm import model_prediction
import pandas as pd
from datetime import datetime 
def prediction(a,b):
    df = pd.DataFrame({
            'Date':pd.date_range(str(datetime.now().date()), periods=b)
        })
    #print(df)
    for i in a:
        #print(i)
        out=model_prediction(i,b)
        #print(out)
        df1 = pd.DataFrame({
            i:out
        })
        df= pd.concat([df, df1], axis = 1)
        #print(df1)
        print(df)
    return df
df=prediction(['Chennai','Bengaluru'],10)
```
